```python
def inclusive_time_by_module(df):
    for dataset in datasets:
        print(dataset)
        dataset_df = df.loc[df['dataset'] == dataset]
        modules = dataset_df['module'].unique()
        for module in modules:
            module_df = dataset_df.loc[dataset_df['module'] == module]['time (inc)'].max()
            print(module, module_df)
        print('======================')    
```

```python
inclusive_time_by_module(df)
```

```python
def exclusive_time_by_module(df):
    for dataset in datasets:
        print(dataset)
        dataset_df = df.loc[df['dataset'] == dataset]
        modules = dataset_df['module'].unique()
        for module in modules:
            module_df = dataset_df.loc[dataset_df['module'] == module]['time']
            print(module, module_df.sum())
        print('======================')
```

```python
exclusive_time_by_module(df)
```

```python
def exclusive_time_by_callsite_and_module(df, dataset):
    df = df.loc[df['dataset'] == dataset]
    callsites = df.loc[df['time'] > 0]['name'].unique()
    sum_of_time = 0
    for callsite in callsites:
        callsite_df = df.loc[df['name'] == callsite]
        max_time = callsite_df['time'].max()
        module = callsite_df['module'].unique()
        sum_of_time += max_time
        print(callsite, module, max_time)
    print(sum_of_time)
```

```python
for dataset in datasets:
    print("Dataset: ", dataset)
    exclusive_time_by_callsite_and_module(df, dataset)
    print("====================")
```

```python
def node_time(module):
    exc_time_sum = 0
    inc_time_max = 0
    module_df = df.loc[df['module'] == module]
    callsites = module_df['name'].unique()
    for callsite in callsites:
        callsite_df = module_name_group_df.get_group((module, callsite))
        max_inc_time = callsite_df['time (inc)'].max()
        inc_time_max = max(inc_time_max, max_inc_time)
        max_exc_time = callsite_df['time'].max()
        exc_time_sum += max_exc_time
#         print(callsite, max_exc_time, max_inc_time)
    return {
        'Exclusive_sum': exc_time_sum,
        'Inclusive_max': inc_time_max
    }
```

```python
for module in modules:
    time_info = node_time(module)
    module_value = max(time_info['Exclusive_sum'], time_info['Inclusive_max'])
    print(module, module_value)
```
