```python
# Global imports
import hatchet as ht
import pandas as pd
import os
```

```python
# Local imports
from callflow import CallFlow
from callflow.operations import ConfigFileReader
```

```python
pd.options.display.max_rows = 10
pd.options.display.float_format = '{:,.2f}'.format
```

```python
path = os.getcwd().split('CallFlow')[0] + 'CallFlow'
```

```python
experiment = 'hpctoolkit-cpi-database'
configFileName = 'config.callflow.json'
```

```python
configPath = path + '/data/' + experiment + '/' + configFileName
```

```python
# Read config file.
config = ConfigFileReader(configPath)
```

```python
# Process single callflow graph. 
callflow = CallFlow(config=config, process=True)
```

```python
# Rendering single callflow graph.
callflow = CallFlow(config=config, process=False)
```

```python
# CCT for single callflow graph.
cct = callflow.request_single({"name": "cct", "dataset": "calc-pi", "functionsInCCT": 50})
print(f"Nodes are: {cct.nodes(data=True)}")
print("\n")
print(f"Edges are: {cct.edges(data=True)}")
```

```python
# Auxiliary data.
auxiliary = callflow.request_single({"name": "auxiliary", "dataset": "calc-pi"})
print(auxiliary)
```

```python
# SuperGraph for single callflow
supergraph = callflow.request_single("supergraph")
```
