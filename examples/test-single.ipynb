```python
import hatchet as ht
import pandas as pd
import os
```

```python
# CallFlow imports
import callflow
from callflow import CallFlow
from callflow.operations import ConfigFileReader
```

```python
pd.options.display.max_rows = 10
pd.options.display.float_format = '{:,.2f}'.format
```

```python
configPath = os.path.abspath("../data/hpctoolkit-cpi-database/callflow.config.json")
```

```python
# Read config file.
config = ConfigFileReader(configPath)
```

```python
dir(callflow)
```

```python
# Process single callflow graph. 
# CallFlow.process (Create the function and copy.)
CallFlow(config=config, process=True)
```

```python
# Rendering single callflow graph.
# Get rid of process here also. 
cf = CallFlow(config=config, ensemble=False)
```

```python
dir(cf)
```

```python
cf.supergraphs
```

```python
dir(cf.supergraphs['calc-pi'].gf)
```

```python
# CCT for single callflow graph.
cct = cf.request_single({"name": "cct", "dataset": "calc-pi", "functionsInCCT": 50})
print(f"Nodes are: {cct.nodes(data=True)}")
print("\n")
print(f"Edges are: {cct.edges(data=True)}")
```

```python
# Auxiliary data.
auxiliary = cf.request_single({"name": "auxiliary", "dataset": "calc-pi"})
print(auxiliary)
```

```python
# SuperGraph for single callflow
supergraph = cf.request_single({"name": "supergraph", "dataset": "calc-pi"})
print(supergraph.edges(data=True))
```
